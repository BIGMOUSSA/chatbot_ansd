```python
import re

def clean_text(input_text):
    pattern = re.compile(r'Tableau.*?Source', re.DOTALL)
    cleaned_text = re.sub(pattern, '', input_text)
    pattern2 = re.compile(r'Graphique.*?Source', re.DOTALL)
    cleaned_text = re.sub(pattern2, '', cleaned_text)
    return cleaned_text

# Example usage:
text_to_clean = "This is some text tableau to be excluded source and more text."
cleaned_text = clean_text(text_to_clean)
print(cleaned_text)
```

```python
def clean_chiffre(input_text):
    pattern = re.compile(r'\d+,\d+,\d+', re.DOTALL)
    cleaned_text = re.sub(pattern, '', input_text)
    #pattern2 = re.compile(r'Graphique.*?Source', re.DOTALL)
    #cleaned_text = re.sub(pattern2, '', cleaned_text)
    return cleaned_text
```

```python
def remove_month_year_occurrences(input_text):
    pattern = re.compile(r'\n(?:janv|fev|mars|avril|mai|juin|juil|août|sept|oct|nov|dec)-\d{2,4}\n')
    cleaned_text = re.sub(pattern, '', input_text)
    pattern2 = re.compile(r'(?:janv|fev|mars|avril|mai|juin|juil|août|sept|oct|nov|dec)-\d{2,4}')
    cleaned_text = re.sub(pattern2, '', cleaned_text)
    pattern3 = re.compile(r'\n(?:janv.|fev.|mars.|avril.|mai.|juin.|juil.|août.|sept.|oct.|nov.|dec.)-\d{2,4}')
    cleaned_text = re.sub(pattern3, '', cleaned_text)
    return cleaned_text
```

```python
def remove_line_break(input_text):
    pattern = re.compile(r'\n')
    cleaned_text = re.sub(pattern, '', input_text)
    return cleaned_text
```

```python
def clean_pub(input_text):
    return remove_line_break(remove_month_year_occurrences(clean_chiffre(clean_text(input_text))))
```

```python
import re

def clean_table_numbers(text):
    # Trouver tous les chiffres entourés par des espaces (possiblement issus des tableaux)
    table_numbers = re.findall(r'\s(\d+)\s', text)

    # Remplacer ces chiffres par une chaîne vide
    cleaned_text = re.sub(r'\s(\d+)\s', ' ', text)

    return cleaned_text, table_numbers

# Exemple d'utilisation
rapport_texte = """
Le tableau ci-dessous montre les résultats :

+----+------+-------+
| ID | Nom  | Score |
+----+------+-------+
| 1  | Jean | 85    |
| 2  | Emma | 92    |
+----+------+-------+

Le score total est 177.
"""

nettoye_texte, chiffres_tableau = clean_table_numbers(rapport_texte)

print("Texte nettoyé :\n", nettoye_texte)
print("\nChiffres du tableau :", chiffres_tableau)
```

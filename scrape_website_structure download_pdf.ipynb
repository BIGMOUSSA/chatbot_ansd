```python
def fetcher(URL):
    assert len(URL) > 0, f"URL expected a non-empty string, got {URL}"
    with requests.Session() as session:
        result = session.get(URL, verify=False)
        result = result.text
        return result
```

```python
def MakeSoupe(URL):
    assert len(URL) > 0, f"URL expected a non-empty string, got {URL}"
    result = fetcher(URL)
    return BeautifulSoup(
        result.encode('utf-8'),
        'html.parser')
```

```python
def getCards(URL):
    assert len(URL) > 0, f"URL expected a non-empty string, got {URL}"
    soupe = MakeSoupe(URL)
    soupe = soupe \
        .find_all("li")
    return soupe
```

```python
year_list = ["2022","2023","2024"]
def contains_year(text, year_list):
    for year in year_list:
        if year in text:
            return True
    return False
```

```python
def get_pdf_link(URL):
    assert len(URL) > 0, f"URL expected a non-empty string, got {URL}"
    soupe = MakeSoupe(URL)
    soupe = soupe \
        .findAll("span", {"class" : "file file--mime-application-pdf file--application-pdf"})
    return soupe
```

```python
import os
# defini une fonction qui prend un dictionnaire
# stock le nom de la publication
# stock le lien
# ouvre le lien
# recupere toutes les publications qui sont sur un interval de temps donnee
# ouvre un dossier portant le nom de la publication
# y stock toutes les publications récuperer en pdf
def MySuperPdfExtractor(data):
    ind = data["indicateurs"].replace("(","").replace(")","").replace(" ","_").replace(":","").replace("/","")
    link = data["source"]
    docs_list = get_pdf_link(link)
    print(f"working with indicateur {ind}")
    if not os.path.exists("Publications_pdf/"+ind): 
        os.mkdir("Publications_pdf/"+ind)
    for item in docs_list:
        if (contains_year(item.text, year_list)) | ("rapport" in str(item.text).lower()) | ("chapitre" in str(item.text).lower()) | ("ses" in str(item.text).lower()) :
            name = str(item.text).replace("(","").replace(")","").replace(" ","_").replace(":","").replace("/","")
            pdf_link = "https://www.ansd.sn/"+item.a["href"]
            response = requests.get(pdf_link, verify=False)
            if response.status_code == 200:
                try : 
                    if not os.path.exists("Publications_pdf/"+ind+"/"+name):
                        with open("Publications_pdf/"+ind+"/"+name, 'wb') as file:
                            file.write(response.content)
                except :
                    try :
                        if not os.path.exists("Publications_pdf/"+name):
                            with open("Publications_pdf/"+name, 'wb') as file:
                                file.write(response.content)
                    except :
                        pass
    print("\n End working \n")
            
            
    
```
